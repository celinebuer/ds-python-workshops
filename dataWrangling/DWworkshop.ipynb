```python
# 9. Add a new column using the data from the current data set & ifelse condition
```

```python
mig_filtered.head()
```

```python
# Q: how can we get the sum of the number of In/Out-migrants for each province/observation?
```

```python
mig_in_out = mig_filtered.pivot(index = ['year', 'month', 'GEO'], columns = 'Interprovincial migration', values = 'count')
```

```python
mig_in_out = mig_in_out.reset_index()
```

```python
mig_in_out.head()
```

```python
# after pivoting, we can take the sum of the last two columns to obtain the total number of migrants per province/observation

mig_in_out['total_migrants'] = mig_in_out['In-migrants'] + mig_in_out['Out-migrants']
```

```python
mig_in_out
```

```python
# q. I want to know for each province/observation whether there was positive influx of migrants or negative

mig_in_out['migrants_flux'] = np.where(mig_in_out['In-migrants']>mig_in_out['Out-migrants'], "positive", "negative")
```

```python
mig_in_out.head()
```

```python
# 10. Derive conclusions from the data set
```

```python
Q: Count the number of positive/negative flux of migrants by each province?

mig_in_out.groupby(['GEO', 'migrants_flux'])['month'].coun()
```
